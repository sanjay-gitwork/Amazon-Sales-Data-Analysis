## Data Cleaning 

* Check for missing or inconsistent data and clean the dataset accordingly.
* Convert price-related columns from string to numerical format.
* Remove unnecessary columns if they do not provide relevant information for your analysis.

```python
df.head(50)
```

```python
df.head(50)
```

```python
df.dtypes
```

```python
#Convert 'ratings' column to float
df['ratings'] = pd.to_numeric(df['ratings'], errors='coerce')

# Convert 'no_of_ratings' column to integer after removing commas
df['no_of_ratings'] = pd.to_numeric(df['no_of_ratings'].str.replace(',', '').astype(float), errors='coerce')

# Convert 'discount_price' and 'actual_price' columns to strings, then remove commas and currency symbols, and finally convert to float
df['discount_price'] = df['discount_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
df.head(50)
```

```python
df.dtypes
```

```python
# Removing the unnecessary columns image and product link
df.drop(['image', 'link'], axis=1, inplace=True)
```

```python
df.head(60)
```

```python
df.tail(60)
```

```python
# Replace missing ratings with a specific placeholder value, like -1, to indicate missing data.
# df['ratings'].fillna(-1, inplace=True)

# Replace missing no_of_ratings with a specific placeholder value, like -1, to indicate missing data.
df['no_of_ratings'].fillna(0, inplace=True)
```

```python
df.head(60)
```

```python
# Step 1: Calculate average ratings for each sub-category
average_ratings_by_subcategory = df.groupby('sub_category')['ratings'].transform('mean').round(2)

print(average_ratings_by_subcategory)

```

```python

# Step 2: Replace NaN values in 'ratings' column with average ratings based on sub-category
df['ratings'].fillna(average_ratings_by_subcategory, inplace=True)
```

```python
df.head(60)
```

```python
# drop rows with NaN values in the 'actual_price' column,
df.dropna(subset=['actual_price'])
```

```python
df.tail()
```

```python
df.head(60)
```

```python
df.tail(60)
```

```python
# replace NaN values in the 'discount_price' column with the corresponding values from the 'actual_price' column where discounts
# are not available. 
df['discount_price'].fillna(df['actual_price'], inplace=True)
```

```python
df.head(60)
```

```python
df.isnull().sum()
```

```python
#  droping rows where both the 'discount_price' and 'actual_price' columns have NaN values ( these are out of stock items)

df.dropna(subset=['discount_price', 'actual_price'], how='all', inplace=True)
```

```python
# Round the 'ratings' column to 2 decimal places
df['ratings'] = df['ratings'].apply(lambda x: round(x, 2))
```

## Exploratory Data Analysis (EDA):

* Visualize Ratings: Create histograms or bar charts to visualize the distribution of ratings.
* Price Analysis: Analyze the distribution of prices for different categories and subcategories.
* Brand Comparison: Compare the average ratings and prices of products from different brands.
* Discount Analysis: Analyze the discounts offered and their impact on sales.
* Most Reviewed Products: Identify the most reviewed products and analyze their characteristics.

```python
# Calculate average discount price and actual price
average_discount_price = df['discount_price'].mean().round(2)
max_discount_price = df['discount_price'].max()
min_discount_price = df['discount_price'].min()


average_actual_price = df['actual_price'].mean().round(2)
max_actual_price = df['actual_price'].max()
min_actual_price = df['actual_price'].min()


average_rating = df['ratings'].mean().round(2)
max_rating = df['ratings'].max()
min_rating = df['ratings'].min()

print(f'Average Discount Price: ₹{average_discount_price:}')
print(f'Maximum Discount Price: ₹{max_discount_price:}')
print(f'Minimum Discount Price: ₹{min_discount_price:}')


print(f'Average Actual Price: ₹{average_actual_price:}')
print(f'Average Ratings: {average_rating:}')
print(f'Max Ratings: {max_rating:}')
print(f'Min Ratings: {min_rating:}')
```

```python
# Plotting a bar chart for ratings
plt.figure(figsize=(20, 6))
df['ratings'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.xlabel('Ratings')
plt.ylabel('Count')
plt.title('Distribution of Ratings')
plt.show()
```

```python
# Plotting a KDE plot for ratings
plt.figure(figsize=(8, 6))
sns.kdeplot(df['ratings'], fill=True, color='skyblue')
plt.xlabel('Ratings')
plt.ylabel('Density')
plt.title('Kernel Density Estimate (KDE) Plot of Ratings')
plt.show()
```
